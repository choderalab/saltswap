# Plotting performance per MD/MC iteration

Loading GPU simulation data from Hal, where each iteraction has been timed. 

```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Image

file = open("hal_gpu.txt").readlines()
time = []
for i in range(3,len(file)):
    time.append(int(file[i][25:28].strip()))
```

Plotting data.

```python
time=np.array(time)
iteration = np.arange(time.shape[0])

plt.scatter(x=iteration,y=time)
plt.xlabel("Iteration")
plt.ylabel("Time (seconds)")
plt.title("Timings for each iteration of MD-MC in SaltSwap")
plt.savefig("IterationTime.png", format='png')
Image("IterationTime.png")
```
