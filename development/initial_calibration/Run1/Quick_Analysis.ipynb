```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import saltswap.analysis_tools as atools
```

## Timeseries plots

```python
potentials = [790, 800, 810, 820, 830, 840]
coords = [(0,0),(0,1),(0,2),(1,0),(1,1),(1,2)]

f, axarr = plt.subplots(2, 3)
for p, c in zip(potentials, coords):
    filename = 'potential_{0}/data.txt'.format(p)
    nwats, nsalt, accprob, time = atools.read_data(filename)
    
    if p == 820:
        col='red'
    else:
        col='C0'
    
    axarr[c].plot(nsalt, color=col)
    axarr[c].set_ylim((0,40))
    axarr[c].set_xlim((0,200))
    axarr[c].set_title('$\Delta\mu$= {} kJ/mol'.format(p), fontsize=14)

## Tidy up the axis labels
plt.setp([a.get_yticklabels() for a in axarr[:, 1]], visible=False) 
plt.setp([a.get_yticklabels() for a in axarr[:, 2]], visible=False)  
plt.setp([a.get_xticklabels() for a in axarr[0, :]], visible=False)  

for a in axarr[:,0]: a.set_ylabel('Number of salt pairs', fontsize=14) 
for a in axarr[1,:]: a.set_xlabel('Iteration', fontsize=14) 

f.set_figheight(8)
f.set_figwidth(10)

plt.tight_layout()
plt.show()
```

The simulations with chemical potentials at 750 and 760 kJ/mol had NCMC GHMC acceptance rates of 0.00, which means the data is extremely untrustworthy. The above figure titled '$\Delta\mu$=820 kJ/mol' has been highlighted in red because it doesn't fit the trend of the data.
